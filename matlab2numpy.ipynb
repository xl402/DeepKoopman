```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt
from mpl_toolkits.mplot3d import Axes3D
```

```python
raw_data_train = genfromtxt('matlab_scripts/FluidFlowOnAttractor_train_x.csv', delimiter=',')
```

```python
raw_data_val = genfromtxt('matlab_scripts/FluidFlowOnAttractor_val_x.csv', delimiter=',')
```

```python
data_train = raw_data_train.reshape((20000, 121, 3))
data_val = raw_data_val.reshape((10000, 121, 3))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for i in np.random.randint(0, len(data_val)-1, 10):
    ax.plot(data_val[i, :, 0], data_val[i, :, 1], data_val[i, :, 2])
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([0, 1])
ax.set_axis_off()
plt.show()
```

```python
np.save('data/fluid_val.npy', data_val)
np.save('data/fluid_train.npy', data_train)
```
